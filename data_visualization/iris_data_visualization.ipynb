```python
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset from sklearn
iris_dataset = load_iris()
iris_df = pd.DataFrame(data=iris_dataset.data, columns=iris_dataset.feature_names)
iris_df['species'] = iris_dataset.target
species_mapping = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
iris_df['species'] = iris_df['species'].map(species_mapping)
```

```python
iris_df.head()
```

```python
# Ensure numerical columns are of the correct data type
iris_df['sepal length (cm)'] = iris_df['sepal length (cm)'].astype(float)
iris_df['sepal width (cm)'] = iris_df['sepal width (cm)'].astype(float)
iris_df['petal length (cm)'] = iris_df['petal length (cm)'].astype(float)
iris_df['petal width (cm)'] = iris_df['petal width (cm)'].astype(float)
```

```python
iris_df.dtypes
```

```python
# Boxplot
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.boxplot(x="species", y="sepal length (cm)", data=iris_df, palette="Set2")
plt.subplot(2, 2, 2)
sns.boxplot(x="species", y="sepal width (cm)", data=iris_df, palette="Set2")
plt.subplot(2, 2, 3)
sns.boxplot(x="species", y="petal length (cm)", data=iris_df, palette="Set2")
plt.subplot(2, 2, 4)
sns.boxplot(x="species", y="petal width (cm)", data=iris_df, palette="Set2")
plt.suptitle("Boxplots of Iris Features by Species", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.histplot(iris_df['sepal length (cm)'], kde=True, color='blue')
plt.title('Distribution of Sepal Length')

plt.subplot(2, 2, 2)
sns.histplot(iris_df['sepal width (cm)'], kde=True, color='orange')
plt.title('Distribution of Sepal Width')

plt.subplot(2, 2, 3)
sns.histplot(iris_df['petal length (cm)'], kde=True, color='green')
plt.title('Distribution of Petal Length')

plt.subplot(2, 2, 4)
sns.histplot(iris_df['petal width (cm)'], kde=True, color='red')
plt.title('Distribution of Petal Width')

plt.tight_layout()
plt.show()
```

```python
# Scatter Plot with Regression Line
plt.figure(figsize=(8, 6))
sns.lmplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=iris_df, aspect=1.5)
plt.title('Sepal Length vs Petal Length with Regression Line')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x="species", y="sepal length (cm)", data=iris_df, palette="Set2", ci="sd")
plt.title('Average Sepal Length by Species with Standard Deviation')
plt.show()
```
